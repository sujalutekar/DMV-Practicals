```python
import pandas as pd

df = pd.read_csv('/content/Housing.csv')

df
```

```python
df.isnull().sum()
```

```python
# Filter and subset the data

filtered_df = df[(df['furnishingstatus'] == 'furnished') & (df['mainroad'] == 'yes') & (df['bedrooms'] >= 2)]

filtered_df
```

```python
# if you want to change the column name

filtered_df.rename(columns = {'bedrooms': 'bedrooms and hall'}, inplace = True)

filtered_df
```

```python
# Clean column names

filtered_df.columns = filtered_df.columns.str.strip().str.replace(' ', '_').str.replace(r'[^a-zA-Z0-9_]', '')

filtered_df
```

```python
#  One-hot encoding
# Example: If a "Location" column has values "Downtown," "Suburb," and "Rural," one-hot encoding will create three new columns: "Location_Downtown," "Location_Suburb," and "Location_Rural."


df_encoded = pd.get_dummies(df, columns=['furnishingstatus', 'basement'], drop_first = True)

df_encoded
```

```python
# Aggregate Data

avg_price_by_area = df.groupby('area')['price'].mean()

avg_price_by_area
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))
sns.boxplot(data = df[['price']])
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR)))]
```

```python
plt.figure(figsize=(20, 8))
sns.boxplot(data = df[['price']])
plt.show()
```
